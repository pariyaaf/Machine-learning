```python
train_data = pd.read_csv("../data/train.csv")
train_data.head(2)
```

```python
test_data = pd.read_csv("../data/test.csv")
test_data.head(10)
```

```python
def count_missing_value(data):
    return ( data == '?').sum().to_dict()


print(count_missing_value(train_data))
        
```

```python
def fill_missing_value(df):
    df = df.copy()
    for col in df.columns:
        max_val = df[df[col] != '?'][col].max()
        
        df.loc[df[col] == '?', col] = max_val
    return df

train_data = fill_missing_value(train_data)

test_data = fill_missing_value(test_data)
```

```python
income = train_data['income']
```

```python
train_data.drop(columns=['race', 'education', 'relationship'], inplace=True)
```

```python
X_train , X_val, y_train, y_val = train_test_split(train_data, income, random_state=42, test_size=0.33)
```

```python
h2o.init()

X = h2o.H2OFrame(X_train)
Y = h2o.H2OFrame(y_train.to_numpy())

X_v = h2o.H2OFrame(X_val)
y_v = h2o.H2OFrame(y_val.to_numpy())

x_test = h2o.H2OFrame(test_data)
```

```python
model = H2ORandomForestEstimator(ntrees=1,
                                 sample_rate=1,
                                 min_rows=20,
                                 categorical_encoding='label_encoder',
                                 balance_classes=True,
                                 nfolds=5,
                                max_depth= 40)

input_columns = [c for c in X.columns if c != 'income']


model.train(
    x=input_columns,
    y='income',
    training_frame=X,
    validation_frame=X_v
)
```

```python
v_pred = model.predict(X_v)
```

```python
y_pred = v_pred['predict'].as_data_frame().values.ravel()
f1 = f1_score(y_val, y_pred , average='binary', pos_label=">50K")
print("F1-score:", f1)
```

```python
y_test = model.predict(x_test)
preds = y_test['predict'].as_data_frame()
```

```python
submission = pd.DataFrame(preds).rename(columns={'predict':'income'})
submission.head()
```

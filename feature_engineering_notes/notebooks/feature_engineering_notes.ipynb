```python
categorical_cols = ['default', 'housing', 'loan', 'is_telephone_contact','job','marital','education', 'month', 'day_of_week', 'poutcome']


for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

df.isna().sum()
```

```python
print(categorical_cols)
continuous_cols = ['age', 'duration', 'pdays', 'previous', 'campaign']
binary_cols = ['default', 'housing', 'loan', 'is_telephone_contact']
nominal_cols = ['job','marital']
ordinal_cols = ['education', 'month', 'day_of_week', 'poutcome']
```

```python
binary_df = df[binary_cols]
binary_df.head()


mapping = {'no': 0,
           'yes' : 1,
           }
for col in binary_cols:
    binary_df[col] = binary_df[col].map(mapping)

binary_df.head()
```

```python
nominal_df =  pd.DataFrame()


for col in nominal_cols:
    print(col)
    dummies = pd.get_dummies(df[col]).rename(columns=lambda x:col+'_' +str(x))
    nominal_df = pd.concat([nominal_df, dummies], axis=1)

 
nominal_df.head()
```

```python
ordinal_df = df[ordinal_cols]
ordinal_df.head()

education_mapping = {
    'illiterate': 0,
    'basic.4y': 1,
    'basic.6y': 2,
    'basic.9y': 3,
    'high.school': 4,
    'professional.course': 5,
    'university.degree': 6
}

day_of_week_mapping = {
    'mon': 0,
    'tue': 1,
    'wed': 2,
    'thu': 3,
    'fri': 4,
    'sat': 5,
    'sun': 6
}

month_mapping = {
    'jan': 0,
    'feb': 1,
    'mar': 2,
    'apr': 3,
    'may': 4,
    'jun': 5,
    'jul': 6,
    'aug': 7,
    'sep': 8,
    'oct': 9,
    'nov': 10,
    'dec': 11
}

poutcome_mapping = {
    'failure': -1,
    'nonexistent': 0,
    'success': 1
}

 



ordinal_df['education'] = ordinal_df['education'].map(education_mapping)
ordinal_df['day_of_week'] = ordinal_df['day_of_week'].map(day_of_week_mapping)
ordinal_df['month'] = ordinal_df['month'].map(month_mapping)
ordinal_df['poutcome'] = ordinal_df['poutcome'].map(poutcome_mapping)


ordinal_df.head()
```

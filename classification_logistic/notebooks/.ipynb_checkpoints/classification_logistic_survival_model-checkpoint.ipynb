```python
train = pd.read_csv('../data/nba_players_train.csv')
test = pd.read_csv('../data/nba_players_test.csv')

train.head(2)
```

```python
test.head(2)
```

```python
test.isnull().sum()
```

```python
train.groupby(['target_5yrs'])['target_5yrs'].count()
```

```python
def preprocessing(data, scaler):
    if 'target_5yrs' in data.columns:
        label = data['target_5yrs']
        data.drop(columns=['target_5yrs'], inplace=True)
    else:
        label = None

    data.drop(columns=['fg', '3p', 'dreb'], inplace=True)
    
    
    data.drop(columns=['name'], inplace=True)
    return scaler.fit_transform(data), label

    
scaler = MinMaxScaler()


train_scale, train_label = preprocessing(train, scaler)
test_scale, _ = preprocessing(test, scaler)

train_scale
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_scale, train_label, test_size=0.33, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


model = LogisticRegression(penalty='l2', fit_intercept=True, class_weight=None, max_iter=300, n_jobs=-1)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score
y_pred = model.predict(X_val)
print('roc= ',roc_auc_score(y_val,y_pred))
```

```python
test_pred = model.predict(test_scale)
submission = pd.DataFrame(test_pred).rename(columns={0:"target_5yrs"})
submission
```

```python
train_data = pd.read_csv('../data/fraud_train.csv')
train_data.head(1)
```

```python
test_data = pd.read_csv('../data/fraud_test.csv')
test_data.head(2)
```

```python
train_data.groupby('Class').count()['Time']
```

```python
label = train_data['Class']
train_data.drop(columns=['Class'], inplace=True)
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_data, label, random_state=42 , test_size = 0.33)
```

```python
# param_grid = {
#     'C' : stats.uniform(2, 10),
#     'gamma': stats.uniform(0.1, 1),
#     'kernel': ['rbf', 'linear','sigmoid']
# }

# model = (SVC(class_weight={1: 10}), param_grid, verbose=3,  n_iter=2)

model = SVC(C=5.336581506738833, gamma=0.3787837386828832, kernel='linear', class_weight={1: 10})
model.fit(X_train, y_train)
```

```python
# print(model.best_estimator_)
```

```python
from sklearn.metrics import f1_score, classification_report
pred_val = model.predict(X_val)

f1_s = f1_score(y_val, pred_val, average = "weighted")
f1_s
```

```python
classification_report(y_val, pred_val)
```

```python
test_pred = model.predict(test_data)
submission = pd.DataFrame(test_pred, columns=['Class'])
submission
```

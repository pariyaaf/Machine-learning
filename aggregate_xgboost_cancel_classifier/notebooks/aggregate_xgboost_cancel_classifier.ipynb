```python
train_data = pd.read_csv('../data/train_data.csv')
train_data.head(2)
```

```python
train_data.groupby('Cancel').count()
```

```python
test_data = pd.read_csv('../data/test_data.csv')
test_data.head(2)
```

```python
train_data.isna().sum()
```

```python
y = train_data['Cancel']

train_data.drop(columns=['Cancel'], inplace=True)
```

```python
train_data.drop(columns=['HashPassportNumber_p','HashEmail','BuyerMobile','NationalCode','UserID', 'VehicleType', 'BillID', 'TicketID'], inplace=True)
```

```python
test_data.drop(columns=['HashPassportNumber_p','HashEmail','BuyerMobile','NationalCode','UserID', 'VehicleType', 'BillID', 'TicketID'], inplace=True)
```

```python
train_data.isna().sum()
```

```python
train_data['VehicleClass'] = train_data['VehicleClass'].fillna(True)
test_data['VehicleClass'] = test_data['VehicleClass'].fillna(True)
```

```python
train_data.columns
```

```python
numerical_columns = ['Created', 'DepartureTime', 'Price', 'CouponDiscount']
categorical_columns = ['ReserveStatus', 'Male', 'From', 'To', 'Domestic', 'VehicleClass',  'Vehicle', 'TripReason']
```

```python
train_data['Created'] = pd.to_datetime(train_data['Created']).astype('int64')
test_data['Created'] = pd.to_datetime(test_data['Created']).astype('int64')

train_data['DepartureTime'] = pd.to_datetime(train_data['DepartureTime']).astype('int64')
test_data['DepartureTime'] = pd.to_datetime(test_data['DepartureTime']).astype('int64')
```

```python
st = StandardScaler()

train_standard = st.fit_transform(train_data[numerical_columns])
test_standard = st.transform(test_data[numerical_columns])

enc = OneHotEncoder(handle_unknown='ignore')

train_enc = enc.fit_transform(train_data[categorical_columns])
test_enc = enc.transform(test_data[categorical_columns])
```

```python
x_train = pd.concat(
    [pd.DataFrame(train_standard), pd.DataFrame(train_enc.toarray())],
    axis=1
)

x_test = pd.concat(
    [pd.DataFrame(test_standard), pd.DataFrame(test_enc.toarray())],
    axis=1
)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(x_train, y, random_state=42, test_size=0.20)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import f1_score
from xgboost import XGBClassifier
import numpy as np

kf = KFold(n_splits=5, shuffle=True, random_state=42)

neg = (y == 0).sum()
pos = (y == 1).sum()
spw = neg / pos

f1_scores = []
best_f1 = -1
best_model = None

for fold, (train_idx, val_idx) in enumerate(kf.split(x_train, y), 1):

    X_tr = x_train.iloc[train_idx]
    X_val = x_train.iloc[val_idx]
    y_tr = y.iloc[train_idx]
    y_val = y.iloc[val_idx]

    model = XGBClassifier(
        n_estimators=300,
        learning_rate=0.1,
        max_depth=50,
        subsample=0.9,
        colsample_bytree=0.9,
        scale_pos_weight=spw,
        eval_metric='logloss',
        n_jobs=-1
    )

    model.fit(X_tr.to_numpy(), y_tr.to_numpy())
    y_pred = model.predict(X_val.to_numpy())

    f1 = f1_score(y_val, y_pred)
    f1_scores.append(f1)

    print(f"Fold {fold} → F1 = {f1:.4f}")
    if f1 > best_f1:
        best_f1 = f1
        best_model = model

print("\nMean F1:", np.mean(f1_scores))
```

```python
val_pred = best_model.predict(X_valid.to_numpy())

print("F1:", f1_score(y_valid, val_pred))
```

```python
y_test = best_model.predict(x_test.to_numpy())
submission = pd.DataFrame(y_test, columns=['Cancel'])
submission
```

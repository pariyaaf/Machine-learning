```python
def find_suggestions(word, n_uniques, search_counts=None, k=5, max_d=3):
    word = str(word)

    suggestions = []
    seen = set()

    for ch in n_uniques:
        if ch.startswith(word) and ch not in seen:
            suggestions.append(ch)
            seen.add(ch)
            if len(suggestions) == k:
                return suggestions

    for d in range(0, max_d + 1):
        for ch in n_uniques:
            if ch in seen:
                continue
            if distance(ch, word) == d:
                suggestions.append(ch)
                seen.add(ch)
                if len(suggestions) == k:
                    return suggestions

    for ch in max_data_repeat:
        if ch not in seen:
            suggestions.append(ch)
            seen.add(ch)
            if len(suggestions) == k:
                break

    return suggestions[:k]
```

```python
class Perceptron:

    def __init__(self):
        self.weights = None

    def weighting(self, x):
        return np.dot(x, self.weights)

    def activation(self, weighted_input):
        return 1 if weighted_input >= 0 else -1

    def predict(self, inputs):
        inputs = np.array(inputs, dtype=float)

        if inputs.ndim == 1:
            inputs = inputs.reshape(1, -1)

        n_samples, n_features = inputs.shape
        inputs_ext = np.hstack([np.ones((n_samples, 1)), inputs])

        preds = []
        for x in inputs_ext:
            z = self.weighting(x)
            preds.append(self.activation(z))

        return np.array(preds)

    def fit(self, inputs, outputs, learning_rate, epochs):
        inputs = np.array(inputs, dtype=float)
        outputs = np.array(outputs)

        if inputs.ndim == 1:
            inputs = inputs.reshape(1, -1)

        n_samples, n_features = inputs.shape
        inputs_ext = np.hstack([np.ones((n_samples, 1)), inputs]) 

        if self.weights is None:
            self.weights = np.random.rand(n_features + 1)

        for epoch in range(epochs):
            for idx, x in enumerate(inputs_ext):
                y_true = outputs[idx]

                z = self.weighting(x)
                y_pred = self.activation(z)

                error = y_true - y_pred

                self.weights = self.weights + learning_rate * error * x

            predicted = self.predict(inputs)
            print(accuracy_score(outputs, predicted))
```

```python
class Adaline:

    def __init__(self):
        self.weights = None

    def weighting(self, x):
        return np.dot(x, self.weights)

    def activation(self, weighted_input):
        return weighted_input

    def predict(self, inputs):
        inputs = np.array(inputs, dtype=float)

        if inputs.ndim == 1:
            inputs = inputs.reshape(1, -1)

        n_samples, n_features = inputs.shape
        inputs_ext = np.hstack([np.ones((n_samples, 1)), inputs])

        preds = []
        for x in inputs_ext:
            z = self.weighting(x)       
            preds.append(1 if z >= 0 else -1)

        return np.array(preds)

    def fit(self, inputs, outputs, learning_rate=0.1, epochs=50):
        inputs = np.array(inputs, dtype=float)
        outputs = np.array(outputs, dtype=float)

        if inputs.ndim == 1:
            inputs = inputs.reshape(1, -1)

        n_samples, n_features = inputs.shape
        inputs_ext = np.hstack([np.ones((n_samples, 1)), inputs])

        if self.weights is None:
            self.weights = np.random.rand(n_features + 1)

        for epoch in range(epochs):
            z = inputs_ext @ self.weights  
            error = outputs - z

            self.weights += learning_rate * (inputs_ext.T @ error)

            predicted = self.predict(inputs)
            print(accuracy_score(outputs, predicted))
```

```python
train = pd.read_csv('../data/yektanet_train.csv')
train.head(2)
```

```python
train.shape
```

```python
test = pd.read_csv('../data/yektanet_test.csv')
test.head(2)
```

```python
train.isna().sum()
```

```python
train['category'].nunique(dropna=True)
```

```python
empty_h1_h2 = train[(train['h1'].isna() | (train['h1'] == "")) &
                    (train['h2'].isna() | (train['h2'] == "")) 
                   ]

empty_h1_h2.shape
```

```python
train['h1'] = train['h1'].fillna(train['title'])
train['h2'] = train['h2'].fillna(train['title'])
train['description'] = train['description'].fillna(train['text_content'])

test['h1'] = test['h1'].fillna(test['title'])
test['h2'] = test['h2'].fillna(test['title'])
test['description'] = test['description'].fillna(test['text_content'])
```

```python
label = train['category']
train.drop(columns=['category', 'url','domain','id', 'text_content'], inplace=True)
test.drop(columns=['url','domain','id', 'text_content'], inplace=True)
```

```python
train.isna().sum()
```

```python
train.head(2)
```

```python
train['all'] = train.apply(lambda r : r['description'] + " " + r['title']+ " " + r['h1']+ " " + r['h2'], axis=1)
test['all'] = test.apply(lambda r : r['description'] + " " + r['title']+ " " + r['h1']+ " " + r['h2'], axis=1)
```

```python
train.head()
```

```python
normalizer = Normalizer()
lemmatizer = Lemmatizer()
stopwords = set(stopwords_list())
pattern_html = re.compile(r"<.*?>", flags=re.DOTALL)
pattern_num = re.compile(r"\d+", flags=re.UNICODE)
pattern_punc = re.compile(r"[^\w\s]")
pattern_underscore = re.compile(r"_+") 

def preprocessing(text):
    if text is None:
        return ""
    try:
        text = normalizer.normalize(text)
    except Exception:
        pass

    text = pattern_html.sub(" ", text)
    text = pattern_underscore.sub(" ", text)
    text = pattern_num.sub(" ", text)
    text = pattern_punc.sub(" ", text)
    
    try:
        text = text.lower()
    except Exception:
        pass
    
    tokens = word_tokenize(text)

    filtered = ''
    for w in tokens:
        if w and w not in stopwords:
            w2 = lemmatizer.lemmatize(w)
            if "#" in w2:
                w2 = w2.split("#")[0]
            filtered = filtered + ' ' + w2
    return filtered.strip()

```

```python
train['all'] = train['all'].apply( lambda r : preprocessing(r))

test['all'] = test['all'].apply( lambda r : preprocessing(r))
```

```python
tfidf_vect = TfidfVectorizer()
tfidf_vect.fit(train['all'])
train_final_counts = tfidf_vect.transform(train['all'])
test_final_counts = tfidf_vect.transform(test['all'])

print(tfidf_vect.get_feature_names_out())
print(train_final_counts.toarray()[0])
print(train_final_counts.toarray()[1])
```

```python
feature_names = tfidf_vect.get_feature_names_out()
train_TF_IDF = pd.DataFrame(train_final_counts.toarray(), columns=feature_names)
train_TF_IDF.head(2)
```

```python
test_TF_IDF = pd.DataFrame(test_final_counts.toarray(), columns=feature_names)
```

```python
pca = PCA(n_components=700)
train_reduced = pca.fit_transform(train_TF_IDF)
test_reduced = pca.transform(test_TF_IDF)
```

```python
test_reduced = pd.DataFrame(test_reduced)
train_reduced = pd.DataFrame(train_reduced)

test_reduced = test_reduced.reset_index(drop=True)
train_reduced = train_reduced.reset_index(drop=True)
```

```python
train_reduced.head(2)
```

```python
lbe = LabelEncoder()
label_lbe = lbe.fit_transform(label)
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_reduced, label_lbe, random_state=42, test_size=0.20)
```

```python
model = XGBClassifier(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.9,
    colsample_bytree=0.9,
    eval_metric='logloss' 
)

model.fit(X_train, y_train)
```

```python
pred = model.predict(X_val)
print("F1:", f1_score(y_val, pred, average='macro'))
```

```python
pred_test = model.predict(test_reduced)

pred_test = list(lbe.inverse_transform(pred_test))
submission = pd.DataFrame({'category': pred_test})

submission
```

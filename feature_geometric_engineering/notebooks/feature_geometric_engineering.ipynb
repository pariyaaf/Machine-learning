```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('../data/train.csv')
df
```

```python
df['label'] = df['outcome'].apply(lambda x: 1 if str(x).strip() in ['گُل', 'گُل به خودی'] else 0)
df.drop(columns=['outcome'], inplace=True)
```

```python
df
# df['label'].value_counts().values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(['matchId', 'playerId', 'playType', 'bodyPart', 'interferenceOnShooter', 'label'], axis=1),
    df['label'],
    random_state=313,
    stratify=df['label']
)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred_proba = model.predict_proba(X_test)[:, 1]
print(f'performance of model is {roc_auc_score(y_test, y_pred_proba)}')
```

```python
df.drop(columns=['playerId', 'matchId'], inplace=True)

df.head()
```

```python
df['bodyPart'] = df['bodyPart'].apply(lambda x : 'پا' if x in ['پای راست','پای چپ'] else x )
df.head(30)
```

```python

df['distance'] = df.apply(lambda i : np.sqrt(i.x**2 + i.y**2), axis=1 )
df.head()
```

```python
df.drop(columns=['x', 'y'], inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
mapping_intrfOnShr = {
    "0": 'کم',
    "1" : 'متوسط'
    }

df['interferenceOnShooter'] = df.apply(
    lambda row : mapping_intrfOnShr.get(str(row.interveningOpponents), 'زیاد')
    if pd.isna(row['interferenceOnShooter'])
    else row ['interferenceOnShooter'],
    axis=1
)
```

```python
df.isna().sum()
```

```python

for col in ['playType', 'bodyPart', 'interferenceOnShooter' ]:
    dummies = pd.get_dummies(df[col]).rename(columns=lambda x:col+'_' +str(x))
    df = pd.concat([df, dummies], axis=1)
```

```python
df.drop(columns=['playType', 'bodyPart', 'interferenceOnShooter'], inplace=True)
```

```python
df.head()
```

```python
y = df['label']
X = df.drop(columns=['label'])
from sklearn.feature_selection import mutual_info_classif
feature_importance = pd.DataFrame(
    mutual_info_classif(X, y, discrete_features=[7, 8], random_state=1401),
    index=X.columns,
    columns=['fi']
).sort_values(by='fi', ascending=False)
```

```python
feature_importance
```

```python
model = LogisticRegression(max_iter=1000)
cols_to_train = feature_importance[feature_importance.fi >= feature_importance.fi.quantile(.5)].index
x = df[cols_to_train]
y = df.label
model.fit(x,y)
x_train,x_test , y_train,y_test = train_test_split(x,y, random_state=313, test_size=.3, stratify=y)
y_pred_proba = model.predict_proba(x_test)[:, 1]
print(f'performance of model is {roc_auc_score(y_test, y_pred_proba)}')
```

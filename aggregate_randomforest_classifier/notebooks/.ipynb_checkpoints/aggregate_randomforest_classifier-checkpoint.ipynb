```python
train_data = pd.read_csv("../data/train.csv")
train_data.head(10)
```

```python
test_data = pd.read_csv("../data/test.csv")
test_data
```

```python
train_data.shape
test_data.shape
```

```python
label = train_data['class']
train_data.drop(columns=['class'], inplace=True)
```

```python
enc = OneHotEncoder(handle_unknown='ignore')

train_enc = enc.fit_transform(train_data)
test_enc = enc.transform(test_data)
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_enc, label, random_state=42, test_size=0.33)
```

```python
X_train.shape
```

```python
model = RandomForestClassifier(n_estimators=200,
                               max_depth=80,
                               bootstrap=True,
                               oob_score=True,
                               max_samples=900)
model.fit(X_train, y_train)
```

```python
val_pred = model.predict(X_val)

recall_score(y_val, val_pred , pos_label='p')
```

```python
y_test = model.predict(test_enc)
prediction = pd.DataFrame(y_test, columns=['class'])
prediction
```

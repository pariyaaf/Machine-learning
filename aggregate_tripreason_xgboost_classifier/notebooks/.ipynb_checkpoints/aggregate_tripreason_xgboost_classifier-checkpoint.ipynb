```python
train_data = pd.read_csv('../data/train_data.csv')
train_data.head(2)
```

```python
# train_data.groupby('TripReason').count()
```

```python
test_data = pd.read_csv('../data/test_data.csv')
test_data.head(2)
```

```python
train_data.isna().sum()
```

```python
train_data.drop(columns=['HashPassportNumber_p','HashEmail','BuyerMobile','NationalCode','CancelTime','UserID', 'VehicleType', 'BillID', 'TicketID'], inplace=True)
```

```python
test_data.drop(columns=['HashPassportNumber_p','HashEmail','BuyerMobile','NationalCode','CancelTime','UserID', 'VehicleType', 'BillID', 'TicketID'], inplace=True)
```

```python
y = train_data['TripReason']
y = y.map({'Work': 0, 'Int': 1})
# y = y.values.ravel()

train_data.drop(columns=['TripReason'], inplace=True)
```

```python
train_data.isna().sum()
```

```python
train_data['VehicleClass'] = train_data['VehicleClass'].fillna(True)
test_data['VehicleClass'] = test_data['VehicleClass'].fillna(True)
```

```python
train_data.head(2)
```

```python
train_data.columns
```

```python
numerical_columns = ['Created', 'DepartureTime', 'Price', 'CouponDiscount']
categorical_columns = ['ReserveStatus', 'Male', 'From', 'To', 'Domestic', 'VehicleClass',  'Vehicle', 'Cancel']
```

```python
train_data['Created'] = pd.to_datetime(train_data['Created']).astype('int64')
test_data['Created'] = pd.to_datetime(test_data['Created']).astype('int64')

train_data['DepartureTime'] = pd.to_datetime(train_data['DepartureTime']).astype('int64')
test_data['DepartureTime'] = pd.to_datetime(test_data['DepartureTime']).astype('int64')
```

```python
st = StandardScaler()

train_standard = st.fit_transform(train_data[numerical_columns])
test_standard = st.transform(test_data[numerical_columns])

enc = OneHotEncoder(handle_unknown='ignore')

train_enc = enc.fit_transform(train_data[categorical_columns])
test_enc = enc.transform(test_data[categorical_columns])
```

```python
x_train = pd.concat(
    [pd.DataFrame(train_standard), pd.DataFrame(train_enc.toarray())],
    axis=1
)

x_test = pd.concat(
    [pd.DataFrame(test_standard), pd.DataFrame(test_enc.toarray())],
    axis=1
)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(x_train, y, random_state=42, test_size=0.33)
```

```python
X_train.shape
```

```python
y_train
```

```python
# !pip install xgboost 
```

```python
model = XGBClassifier(
    n_estimators=200,
    learning_rate= 0.05,
    max_depth=12,
    susample=0.9,
    colsample_bytree=0.9,
    eval_metric='logloss'
)
```

```python
model.fit(X_train.to_numpy(), y_train)
```

```python
val_pred = model.predict(X_valid.to_numpy())

print("F1:", f1_score(y_valid, val_pred))
```

```python
y_test = model.predict(x_test.to_numpy())
y_test = pd.Series(y_test)
y_test = y_test.map({0: 'Work', 1: 'Int'})
submission = pd.DataFrame(y_test, columns=['TripReason'])
submission
```
